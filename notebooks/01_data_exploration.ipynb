# Đọc và Phân Tích Dữ Liệu CSV

Notebook này giúp bạn đọc và hiển thị dữ liệu CSV theo cách dễ nhìn

## 1. Import thư viện cần thiết

```python
import pandas as pd
import os

# Cấu hình pandas để hiển thị đầy đủ
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 100)
```

## 2. Đọc file CSV - train.csv

```python
# Đọc file CSV
df_train = pd.read_csv('data/raw/train.csv')

# Hiển thị thông tin cơ bản
print(f"Kích thước dữ liệu: {df_train.shape [0]} dòng × {df_train.shape[1]} cột")
print(f"\nTên các cột:")
print(df_train.columns.tolist())
```

## 3. Xem 10 dòng đầu tiên

```python
df_train.head(10)   
```

## 4. Thông tin chi tiết về dữ liệu

```python
# Hiển thị thông tin chi tiết
df_train.info()
```

## 5. Thống kê mô tả

```python
# Thống kê cho các cột số
df_train.describe()
```

## 6. Lọc và xem dữ liệu cụ thể

```python
# Xem các comment có rating 5 sao
high_rating = df_train[df_train['n_star'] == 5].head(5)
print("Các comment có 5 sao:")
print(high_rating[['comment', 'n_star', 'label']].to_string())
```

```python
# Xem các comment có rating 1 sao
low_rating = df_train[df_train['n_star'] == 1].head(5)
print("Các comment có 1 sao:")
print(low_rating[['comment', 'n_star', 'label']].to_string())
```

## 7. Phân tích các nhãn (label)

```python
# Đếm số comment theo rating
print("Phân bố rating:")
print(df_train['n_star'].value_counts().sort_index())

print(f"\nTổng số label duy nhất: {df_train['label'].nunique()}")
```

## 8. Đọc các file khác

```python
# Đọc file dev.csv
df_dev = pd.read_csv('data/raw/dev.csv')
print(f"Dev set: {df_dev.shape[0]} dòng × {df_dev.shape[1]} cột")
print(df_dev.head(3))
```

```python
# Đọc file test.csv
df_test = pd.read_csv('data/raw/test.csv')
print(f"Test set: {df_test.shape[0]} dòng × {df_test.shape[1]} cột")
print(df_test.head(3))
```
